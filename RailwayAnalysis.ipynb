```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## India's Railway Accident Data Analysis - 2019

#### Loading Dataset

```python
df = pd.read_csv('Railway2019.csv')
print(f"Dataset loaded successfully")
print(f"shape:{df.shape}")
```

#### Data Exploration

```python
df.head()
```

```python
print("Dataset info:")
df.info()
```

```python
print("Basic Statistics:")
df.describe()
```

#### Data Cleaning

```python
print("Missing values per column:")
df.isnull().sum()
```

```python
# CLeaning Column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df_clean = df[~df['State/UT'].str.contains('TOTAL',na=False)].copy()
df_clean
```

```python
print(f"Original dataset: {len(df)} rows")
print(f"After removing rows with 'TOTAL' total rows: {len(df_clean)} rows")
```

#### Data Filtering and Grouping

```python
states_df = df_clean[df_clean['Category'] == 'State'].copy()
uts_df = df_clean[df_clean['Category'] == 'UT'].copy()
print(f"States: {len(states_df)}")
print(f"Union Territories: {len(uts_df)}")
```

```python
#creating accidnet type columns for easier analysis 
accident_types = ['Collisions','Derailments','Fall from Train/Collision with People at Tracks','Explosion/Fire','Others']
```

#### Aggregations

```python
#Prforming Aggregations
#Top 10 states by total deaths
top_deaths = states_df.nlargest(10,'Total_Died')[['State/UT','Total_Died','Total_Cases','Total_Injured']]
print("Top 10 States by Total Deaths:")
print(top_deaths)
```

```python
#Top 10 States bby total cases
top_cases = states_df.nlargest(10,'Total_Cases')[['State/UT','Total_Died','Total_Cases','Total_Injured']]
print("Top 10 States by Total Cases:")
print(top_cases)
```

```python
#Aggregate by accident type
accident_summary = pd.DataFrame({
    'Accident_Type': accident_types,
    'Total_Cases': [
        states_df['Collisions_Cases'].sum(),
        states_df['Derailments_Cases'].sum(),
        states_df['Fall from Train/Collision with People at Tracks_Cases'].sum(),
        states_df['Explosion/Fire_Cases'].sum(),
        states_df['Others_Cases'].sum()
    ],
    'Total_Deaths': [
        states_df['Collisions_Died'].sum(),
        states_df['Derailments_Died'].sum(),
        states_df['Fall from Train/Collision with People at Tracks_Died'].sum(),
        states_df['Explosion/Fire_Died'].sum(),
        states_df['Others_Died'].sum()
    ],
    'Total_Injured': [
        states_df['Collisions_Injured'].sum(),
        states_df['Derailments_Injured'].sum(),
        states_df['Fall from Train/Collision with People at Tracks_Injured'].sum(),
        states_df['Explosion/Fire_Injured'].sum(),
        states_df['Others_Injured'].sum()
    ]
})
print("Accident summary by Type:")
print(accident_summary)
```

#### Data Visualization

```python
fig = plt.figure(figsize=(20,15))
```

```python
# Top 10 States by Total Deaths
top_10_deaths = states_df.nlargest(10, 'Total_Died')
bars1 = plt.bar(range(len(top_10_deaths)), top_10_deaths['Total_Died'], color='red', alpha=0.7)
plt.title('Top 10 States by Total Deaths', fontsize=14, fontweight='bold')
plt.xlabel('States')
plt.ylabel('Total Deaths')
plt.xticks(range(len(top_10_deaths)), top_10_deaths['State/UT'], rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.savefig('total_deaths.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
# 2. Top 10 States by Total Cases
top_10_cases = states_df.nlargest(10, 'Total_Cases')
bars2 = plt.bar(range(len(top_10_cases)), top_10_cases['Total_Cases'], color='orange', alpha=0.7)
plt.title('Top 10 States by Total Cases', fontsize=14, fontweight='bold')
plt.xlabel('States')
plt.ylabel('Total Cases')
plt.xticks(range(len(top_10_cases)), top_10_cases['State/UT'], rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.savefig('total_cases.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases.png'")
```

```python
#Injuries Vs Deaths Comparison
plt.figure(figsize=(14,6))
df_sorted = df.sort_values("Total_Died",ascending = False)
plt.bar(df_sorted["State/UT"], df_sorted["Total_Died"], label="Deaths", color='crimson')
plt.bar(df_sorted["State/UT"], df_sorted["Total_Injured"], label="Injured", bottom=df_sorted["Total_Died"], color='orange')
plt.xticks(rotation=90)
plt.xlabel("State/UT")
plt.ylabel("Count")
plt.title("Comparison of Injuries and Deaths by State")
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('injuries vs deaths.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
# 3. Accident Types Distribution (Pie Chart)
accident_cases = accident_summary['Total_Cases']
accident_labels = [label.replace('Fall from Train/Collision with People at Tracks', 'Track Accidents') 
                   for label in accident_summary['Accident_Type']]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
wedges, texts, autotexts = plt.pie(accident_cases, labels=accident_labels, autopct='%1.1f%%', 
                                   colors=colors, startangle=90)
plt.title('Distribution of Accident Types by Cases', fontsize=14, fontweight='bold')
plt.savefig('accident type pie.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
# 5. Accident Type Comparison (Deaths)

accident_deaths = accident_summary['Total_Deaths']
bars5 = plt.bar(range(len(accident_summary)), accident_deaths, color='darkred', alpha=0.7)
plt.title('Deaths by Accident Type', fontsize=14, fontweight='bold')
plt.xlabel('Accident Type')
plt.ylabel('Total Deaths')
plt.xticks(range(len(accident_summary)), 
           [label.replace('Fall from Train/Collision with People at Tracks', 'Track\nAccidents') 
            for label in accident_summary['Accident_Type']], rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.savefig('death by acc type.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
# Define risk levels based on total deaths
def classify_risk(deaths):
    if deaths >= 1000:
        return 'High Risk'
    elif deaths >= 100:
        return 'Moderate Risk'
    elif deaths > 0:
        return 'Low Risk'
    else:
        return 'No Risk'

# Apply classification
df['Risk_Level'] = df['Total_Died'].apply(classify_risk)

# Count number of states in each risk level
risk_distribution = df['Risk_Level'].value_counts()
risk_group = df[['State/UT', 'Risk_Level']].groupby('Risk_Level')['State/UT'].apply(list)
print(risk_group)


# Pie chart
colors = ['red', 'orange', 'yellow', 'green']
plt.figure(figsize=(8, 6))
plt.pie(risk_distribution, labels=risk_distribution.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Distribution of States by Risk Level (Based on Total Deaths in 2019)', fontsize=12)
plt.axis('equal')
plt.tight_layout()
plt.show()
plt.savefig('pie chart.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
#Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
print("Plot saved as 'total_cases_by_state_bar_plot.png'")
```

```python
# 7. SUMMARY STATISTICS
print("\n7. Summary Statistics:")
print("=" * 30)

# Overall statistics
total_cases = states_df['Total_Cases'].sum()
total_deaths = states_df['Total_Died'].sum()
total_injured = states_df['Total_Injured'].sum()
overall_fatality_rate = (total_deaths / total_cases * 100) if total_cases > 0 else 0

print(f"Total Cases: {total_cases:,}")
print(f"Total Deaths: {total_deaths:,}")
print(f"Total Injured: {total_injured:,}")
print(f"Overall Fatality Rate: {overall_fatality_rate:.2f}%")

print(f"\nMost Dangerous State (by deaths): {top_deaths.iloc[0]['State/UT']} ({top_deaths.iloc[0]['Total_Died']} deaths)")
print(f"Highest Case Count: {top_cases.iloc[0]['State/UT']} ({top_cases.iloc[0]['Total_Cases']} cases)")

# Most common accident type
most_common_accident = accident_summary.loc[accident_summary['Total_Cases'].idxmax()]
print(f"Most Common Accident Type: {most_common_accident['Accident_Type']} ({most_common_accident['Total_Cases']} cases)")

# Deadliest accident type
deadliest_accident = accident_summary.loc[accident_summary['Total_Deaths'].idxmax()]
print(f"Deadliest Accident Type: {deadliest_accident['Accident_Type']} ({deadliest_accident['Total_Deaths']} deaths)")

print("\nAnalysis Complete!")
```
